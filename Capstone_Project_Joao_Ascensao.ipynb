# Data Transformation

* Just as good practice, we will create a copy of the dataset to start transforming and cleaning the data. Keeping the original dataset untouched.

```python
#Do a copy of the dataset to proceed with the data transformation
ds2 = ds.copy()
```

```python
#Check Missing values- create table
def missing_values_table(x):
        mv = x.isna().sum()
        mv_pct = 100 * x.isna().sum() / len(x)
        mv_table = pd.concat([mv, mv_pct], axis=1)
        mv_table = mv_table.rename(
        columns = {0: 'N of Missing Values', 1: '% of Missing Values'})
        mv_table = mv_table[
            mv_table.iloc[:,1] != 0].sort_values(
        '% of Missing Values', ascending=False).round(1)
        return mv_table

missing_values_table(ds2)
```

```python
# drop rows with missing values in 'Price' column
ds2.dropna(subset=['Price'], inplace=True)

# drop 'New_price' column as it has too many missing values
ds2.drop(['New_price'], axis=1, inplace=True)
```

* To start cleaning the data, we drop the rows with missing values in the 'price' variable, because it is our target variable and for that reason we do not want create any bias in this feature by imputing new values with the mean or median.
* We drop the 'New_Price' column entirely  for the reasons already mention. 


```python
# fill missing values in 'Mileage' column with median value
median_mileage = ds2['Mileage'].median()
ds2['Mileage'].fillna(median_mileage, inplace=True)

# fill missing values in 'Engine' column with median value
median_engine = ds2['Engine'].median()
ds2['Engine'].fillna(median_engine, inplace=True)

# fill missing values in 'Power' column with median value
median_power = ds2['Power'].median()
ds2['Power'].fillna(median_power, inplace=True)

# fill missing values in 'Seats' column with mode value
mode_seats = ds2['Seats'].mode()[0]
ds2['Seats'].fillna(mode_seats, inplace=True)
```

* For the variables 'Mileage', 'Engine', and 'Power', we replace the missing values with the median values because it's generally better not to lose data. Being a good practice to do imputation instead of drop if the percentage of missing values in the variable is insignificant. Median because it tends to create less bias in the data then the mean.
* For the variable 'Seats' we impute with the mode because it is a discrete variable.

```python
#Check that there aren't any Missing values anymore 
def missing_values_table(x):
        mv = x.isna().sum()
        mv_pct = 100 * x.isna().sum() / len(x)
        mv_table = pd.concat([mv, mv_pct], axis=1)
        mv_table = mv_table.rename(
        columns = {0: 'N of Missing Values', 1: '% of Missing Values'})
        mv_table = mv_table[
            mv_table.iloc[:,1] != 0].sort_values(
        '% of Missing Values', ascending=False).round(1)
        return mv_table

missing_values_table(ds2)
```

## Lasso Regression Model Evaluation

```python
# Define a function to calculate and return a dataframe containing performance metrics for train and test sets
def performanceMetricsDF(yTrain, yPredTrain, yTest, yPredTest):
    measures_list = ['MAE', 'RMSE', 'R^2', 'MAX Error']
    train_results = [metrics.mean_absolute_error(yTrain, yPredTrain),
                     np.sqrt(metrics.mean_squared_error(yTrain, yPredTrain)),
                     metrics.r2_score(yTrain, yPredTrain),
                     metrics.max_error(yTrain, yPredTrain)]
    test_results = [metrics.mean_absolute_error(yTest, yPredTest),
                    np.sqrt(metrics.mean_squared_error(yTest, yPredTest)),
                    metrics.r2_score(yTest, yPredTest),
                    metrics.max_error(yTest, yPredTest)]
    resultsDF = pd.DataFrame({'Measure': measures_list, 'Train': train_results, 'Test': test_results})
    return resultsDF

# Define a function to calculate and print the performance metrics of a model
def get_model_score(model, X_train, y_train, X_test, y_test, flag=True):
    score_list = []
    
    # Make predictions on the training and test data
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)

    # Convert the predicted values back to the original scale
    pred_train = np.exp(pred_train)
    pred_test = np.exp(pred_test)
    
    # Calculate the performance metrics for the training and test data
    perf_metrics_df = performanceMetricsDF(y_train['Price'], pred_train, y_test['Price'], pred_test)
    
    # Calculate the R-squared and RMSE scores for the training and test data
    train_r2 = metrics.r2_score(y_train['Price'], pred_train)
    test_r2 = metrics.r2_score(y_test['Price'], pred_test)
    train_rmse = metrics.mean_squared_error(y_train['Price'], pred_train, squared=False)
    test_rmse = metrics.mean_squared_error(y_test['Price'], pred_test, squared=False)

    score_list.extend((train_r2, test_r2, train_rmse, test_rmse))

    if flag == True:
        print("R-squared on training set: {:.3f}".format(train_r2))
        print("R-squared on test set: {:.3f}".format(test_r2))
        print("RMSE on training set: {:.3f}".format(train_rmse))
        print("RMSE on test set: {:.3f}".format(test_rmse))
        print(perf_metrics_df)
        
    return score_list

# Call the get_model_score function with the Lasso regression model
model_score = get_model_score(lasso_regr, X_train, y_train, X_test, y_test, flag=True)
```

## Random Forest Regression Model Evaluation

```python
def get_model_score(model, X_train, y_train, X_test, y_test, flag=True):
    score_list = []

    # Make predictions on the training and test data
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)

    # Convert the predicted values back to the original scale
    pred_train = np.exp(pred_train)
    pred_test = np.exp(pred_test)

    # Calculate the R-squared and RMSE scores for the training and test data
    train_r2 = metrics.r2_score(y_train['Price'], pred_train)
    test_r2 = metrics.r2_score(y_test['Price'], pred_test)
    train_rmse = metrics.mean_squared_error(y_train['Price'], pred_train, squared=False)
    test_rmse = metrics.mean_squared_error(y_test['Price'], pred_test, squared=False)

    score_list.extend((train_r2, test_r2, train_rmse, test_rmse))

    if flag == True:
        print("R-squared on training set: {:.3f}".format(train_r2))
        print("R-squared on test set: {:.3f}".format(test_r2))
        print("RMSE on training set: {:.3f}".format(train_rmse))
        print("RMSE on test set: {:.3f}".format(test_rmse))

    return score_list

# Evaluate the original RandomForestRegressor model
print("Original RandomForestRegressor model:")
original_scores = get_model_score(rf_regressor, X_train, y_train, X_test, y_test)

# Evaluate the best RandomForestRegressor model
print("\nBest RandomForestRegressor model:")
best_scores = get_model_score(best_rf_model, X_train, y_train, X_test, y_test)

# Create a comparison table
comparison_table = pd.DataFrame(
    {
        'Metric': ['R-squared (Train)', 'R-squared (Test)', 'RMSE (Train)', 'RMSE (Test)'],
        'Original Model': original_scores,
        'Best Model': best_scores,
    }
)
comparison_table
```
